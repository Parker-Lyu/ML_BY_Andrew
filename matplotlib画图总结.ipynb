## 三维散点图

```python
X, y = make_circles(100, factor=.6, noise=.1)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn')
plt.show()
```

```python
r = np.exp(-(X ** 2).sum(1))
def plot_3D(elev=30, azim=30, X=X, y=y):
    plt.figure(dpi=100)
    ax = plt.subplot(projection='3d')
    ax.scatter3D(X[:, 0], X[:, 1], r, c=y, s=50, cmap='autumn')
    ax.view_init(elev=elev, azim=azim)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('r')

plot_3D(elev=45, azim=45, X=X, y=y)
```

```python
# 另一种画图方法
fig = plt.figure(1)
ax = Axes3D(fig)
ax.set_xlabel('X label',color='r') #设置x坐标
ax.set_ylabel('Y label',color='r')
ax.set_zlabel('Z label')

r = np.exp(-(X ** 2).sum(1))
ax.scatter3D(X[:,0],X[:,1], r,c=y,cmap=plt.get_cmap('rainbow'))#生成一个曲面
plt.show()
```

```python
# 从4维数据中挑出三维画图
np.random.seed(5)
iris = datasets.load_iris()
X = iris.data
y = iris.target

fig = plt.figure( figsize=(8,6))
ax = Axes3D(fig, rect=[0,0,.95,1], elev = 48, azim=134)

# 在三维图中写出文字
for name, label in [('Setosa', 0), ('Versicolour', 1), ('Virginica', 2)]:
    ax.text3D(X[y==label, 3].mean(),
             X[y==label, 0].mean(),
             X[y==label, 2].mean() + 2, name, horizontalalignment='center',
              bbox = dict(alpha=.2, edgecolor='w', facecolor='w'))

ax.scatter(X[:,3], X[:,0], X[:,2], c=y, edgecolor='k', cmap=plt.get_cmap('rainbow'))
ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])
ax.set_xlabel('Petal width')
ax.set_ylabel('Sepal length')
ax.set_zlabel('Petal length')
ax.set_title('Ground Truth')
ax.dist = 12
plt.show()
```

# 三维表面图，并在三维图上画等高线图
原理：生成平面上的网格数据，即生成XOY平面上的每一点，再对这些坐标对进行映射，生成对应的z值，这样，xyz的坐标就有了，可以用来画面了

```python
fig = plt.figure(dpi=100)
ax = Axes3D(fig)     #生成一个3d对象
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)    #对X,Y数组进行扩充
R = np.sqrt(X**2+Y**2)
Z = np.sin(R)
ax.set_xlabel('X label',color='r') #设置x坐标
ax.set_ylabel('Y label',color='r')
ax.set_zlabel('Z label')
ax.set_zlim(-2,2)
ax.plot_surface(X, Y, Z,rstride=1,cstride=1,cmap=plt.cm.hot)#生成一个曲面   cmap = plt.get_cmap('rainbow')
ax.contourf(X,Y,Z,offset=-2,alpha=0.75,cmap=plt.cm.hot)#为等高线填充颜色
# ax.contour(X, Y, Z, offset=-2)#生成等高线 offset参数是等高线所处的位置
```

# 关于legend

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.covariance import EmpiricalCovariance, MinCovDet
```

```python
# 参数
n_samples = 125
n_outliers = 25
n_features = 2

# 构建协方差矩阵
gen_cov = np.eye(n_features)
gen_cov[0,0] = 2
# 生成按上面协方差矩阵分布的数据
X = np.random.randn(n_samples, n_features).dot(gen_cov)

# 离群点与正样本点服从不同的高斯分布
outliers_cov = np.eye(n_features)
outliers_cov[np.arange(1,n_features),np.arange(1,n_features)] = 7.
X[-n_outliers:] = np.random.randn(n_outliers, n_features).dot(outliers_cov)
```

```python
# fit a Minimum Covariance Determinant (MCD) robust estimator to data
robust_cov = MinCovDet().fit(X)

# compare estimators learnt from the full data set with true parameters
emp_cov = EmpiricalCovariance().fit(X)
```

```python
# 点数据
fig = plt.figure(figsize=(10,8))
inlier_plot = plt.scatter(X[:-n_outliers][:,0], X[:-n_outliers][:,1], c='b')
outlier_plot = plt.scatter(X[-n_outliers:][:,0], X[-n_outliers:][:,1], c='r')


xx,yy = np.meshgrid(np.linspace(X[:,0].min(),X[:,0].max(),100),np.linspace(X[:,1].min(),X[:,1].max(),100))
zz = np.c_[xx.ravel(),yy.ravel()]
#最大似然协方差估计
mahal_emp_cov = emp_cov.mahalanobis(zz)
print(mahal_emp_cov.shape)
mahal_emp_cov = mahal_emp_cov.reshape(xx.shape)
print(mahal_emp_cov.shape)
emp_cov_contour = plt.contour(xx, yy, np.sqrt(mahal_emp_cov),
                                  cmap=plt.cm.PuBu_r,
                                  linestyles='dashed')
plt.clabel(emp_cov_contour,inline=True,fontsize=10) #在等高线上标出对应的z值

# 稳健协方差估计
mahal_robust_cov = robust_cov.mahalanobis(zz)
mahal_robust_cov = mahal_robust_cov.reshape(xx.shape)
robust_contour = plt.contour(xx, yy, np.sqrt(mahal_robust_cov),
                                 cmap=plt.cm.YlOrBr_r, linestyles='dotted')
plt.clabel(robust_contour,inline=True,fontsize=10) #在等高线上标出对应的z值

plt.legend([emp_cov_contour.collections[1], robust_contour.collections[1],
                inlier_plot, outlier_plot],
          ['MLE dist', 'robust dist', 'inliers', 'outliers'],
          loc="upper right", borderaxespad=0)

plt.show()
```

# 隔离森林画图

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt
```

```python
rng = np.random.RandomState(42)
```

```python
# 生成数据
X = 0.3 * rng.randn(100,2)
X_train = np.r_[X+2, X-2]
X = 0.3 * rng.randn(25,2).dot(np.array([[3,0],[0,3]]))
X_test = np.r_[X+2,X-2]
X_outliers = rng.uniform(low=-6, high=6, size=(20,2))
data = np.r_[X_train,X_test,X_outliers]

plt.figure(figsize=(10,8))
plt.scatter(data[:200][:,0], data[:200][:,1], label='data1',s=10)
plt.scatter(data[200:250][:,0],data[200:250][:,1], label='data2',s=10)
plt.scatter(data[250:][:,0], data[250:][:,1], label='outliers',s=10)
xlim = plt.xlim()
ylim = plt.ylim()
plt.legend()
plt.show()
print(xlim)
print(ylim)
```

```python
clf = IsolationForest(random_state=rng)
clf.fit(data)
# y_pred_train = clf.predict(X_train)
# y_pred_test = clf.predict(X_test)
# y_pred_outliers = clf.predict(X_outliers)

xx,yy = np.meshgrid(np.linspace(xlim[0],xlim[1],50),np.linspace(ylim[0],ylim[1],50))
print(xx.shape)
Z = clf.decision_function(np.c_[xx.ravel(),yy.ravel()])
print(Z.shape)
Z = Z.reshape(xx.shape)
print(Z.shape)

plt.figure(figsize=(10,8))
forest = plt.contourf(xx, yy, Z, cmap=plt.cm.Blues_r,)
b1 = plt.scatter(data[:200][:,0], data[:200][:,1], c='white',label='data1',s=20,
                 edgecolor='k')
b2 = plt.scatter(data[200:250][:,0], data[200:250][:,1], c='green',label='data2',
                 s=20, edgecolor='k')
c = plt.scatter(data[250:][:,0], data[250:][:,1], c='red',label='outlier',
                s=20, edgecolor='k')
forest
plt.legend()
plt.show()
```
