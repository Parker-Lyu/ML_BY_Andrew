```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python

fig = plt.figure(1)
ax = Axes3D(fig)     #生成一个3d对象
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)    #对X,Y数组进行扩充
R = np.sqrt(X**2+Y**2)
Z = np.sin(R)
ax.set_xlabel('X label',color='r') #设置x坐标
ax.set_ylabel('Y label',color='r')
ax.set_zlabel('Z label')
ax.plot_surface(X, Y, Z,rstride=1,cstride=1,cmap=plt.cm.hot)#生成一个曲面   cmap = plt.get_cmap('rainbow')
#ax.contourf(X,Y,Z,offset=2,alpha=0.75,cmap=plt.cm.hot)#为等高线填充颜色
ax.contour(X, Y, Z, offset=-2, colors='black')#生成等高线 offset参数是等高线所处的位置
 
fig=plt.figure(2)
 
bx=fig.gca()#再生成一个字图
 
C=bx.contour(X,Y,Z) #如果想要在等高线上标出相应的值，需要重新生成一个对象，不能是3d对象
bx.clabel(C,inline=True,fontsize=10) #在等高线上标出对应的z值
ax.set_zlim(-2,2)#设置z的范围
plt.show()
```

```python
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
import scipy as sp
```

```python
iris = datasets.load_iris().data[:,1]
iris
```

```python
bins, scaler = np.histogram(iris, bins=20)
print(bins)
print(scaler)
```

```python
scaler_deals = (scaler[1:] + scaler[:-1]) * 0.5
plt.bar(scaler_deals, bins,edgecolor='b',width=0.1)
plt.show()
```

```python
face = sp.misc.face()
plt.imshow(face)
plt.show()
```

```python
color_ = ['r','g','b']
for i in range(3):
    channel = face[:,:,i].flatten()
    fig = plt.figure()
    plt.hist(channel,bins=256,color=color_[i],hold=1)
plt.show()
    
```
