```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python

fig = plt.figure(1)
ax = Axes3D(fig)     #3d
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)    #X,Y
R = np.sqrt(X**2+Y**2)
Z = np.sin(R)
ax.set_xlabel('X label',color='r') #x
ax.set_ylabel('Y label',color='r')
ax.set_zlabel('Z label')
ax.plot_surface(X, Y, Z,rstride=1,cstride=1,cmap=plt.cm.hot)#
#ax.contourf(X,Y,Z,offset=2,alpha=0.75,cmap=plt.cm.hot)#
ax.contour(X, Y, Z, offset=-2, colors='black')# offset
 
fig=plt.figure(2)
 
bx=fig.gca()#
 
C=bx.contour(X,Y,Z) #3d
bx.clabel(C,inline=True,fontsize=10) #z
ax.set_zlim(-2,2)#z
plt.show()
```
